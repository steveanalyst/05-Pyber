## Pyber Data Analysis

Observations:
1. Based on "Pyber Ride Sharing Data 2018", Pyber has more dirvers in Urban area but the average fare is the lowest among all three different market sections (Urban, Suburban and Rural). It could be related with shorter distance for each ride.
2. Based on Pie Chart "Total Fares by City Type" and "% ot Total Drivers by City Type", Suburban has contributed 30.5% fares, which is a pretty strong contribution to the total fare income, considering Suburban driver count is only 16.5%. This makes Suburban to be the most profitable market.
3. Urban areas have most of drivers and Suburban area has more growing spaces.

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "/Users/wheat/Pyber/data/city_data.csv"
ride_data_to_load = "/Users/wheat/Pyber/data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df=pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
combined_df = pd.merge(ride_df,city_df,  on='city',how='left')

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#Using Loc method to get different type city list dataframe
urban_df = combined_df[combined_df["type"] == "Urban"].groupby([combined_df["city"]])
suburban_df = combined_df[combined_df["type"] == "Suburban"].groupby([combined_df["city"]])
rural_df = combined_df[combined_df["type"] == "Rural"].groupby([combined_df["city"]])

#Get each city's ride count
urban_ride_count = urban_df['ride_id'].count()
suburban_ride_count = suburban_df['ride_id'].count()
rural_ride_count = rural_df['ride_id'].count()

#Get each city's driver count
urban_city_driver_count = urban_df['driver_count'].mean()
suburban_city_driver_count = suburban_df['driver_count'].mean()
rural_city_driver_count = rural_df['driver_count'].mean()

#Get average fare
urban_avg_fare = urban_df['fare'].mean()
suburban_avg_fare = suburban_df['fare'].mean()
rural_avg_fare = rural_df['fare'].mean() 

# Build the scatter plots for each city types
plt.scatter(urban_ride_count, urban_avg_fare, s=urban_city_driver_count*10,
                           marker="o", facecolors="coral", edgecolors="black",alpha=.75, label="Urban")
plt.scatter(suburban_ride_count, suburban_avg_fare, s=suburban_city_driver_count*10,
                              marker="o",facecolors="lightskyblue", edgecolors="black",alpha=.75, label="Suburban")
plt.scatter(rural_ride_count, rural_avg_fare, s=rural_city_driver_count*10,
                           marker="o",facecolors="gold", edgecolors="black",alpha=.75,label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data 2018")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
legend = plt.legend(fontsize = 8,  title="City Types", loc="best") 


# Adjust the legend 
legend.legendHandles[0]._sizes = [35]
legend.legendHandles[1]._sizes = [35]
legend.legendHandles[2]._sizes = [35]



# Incorporate a text label regarding circle size
plt.text(42, 35, "Note: \nCircle size correlates with driver count per city.", fontsize=10)

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/PyberGraph.png", bbox_inches="tight")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
types=combined_df.groupby(['type'])
total_fare = types['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,.1)

plt.pie(total_fare, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/PercentageOfTotalFaresByCityType.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride = types['ride_id'].count()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,.1)
plt.pie(total_ride, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/PctOfTotalRidesByCityType.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_type_group= city_df.groupby(['type'])
driver_count_citytype= city_type_group['driver_count'].sum()

# Build Pie Charts
labels = ["Rural","Suburban","Urban"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0,.1)
plt.pie(driver_count_citytype, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/TotalDriversbyCityType.png")

# Show Figure
plt.show()
```

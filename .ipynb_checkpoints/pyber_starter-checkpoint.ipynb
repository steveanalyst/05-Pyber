```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "/Users/wheat/Pyber/data/city_data.csv"
ride_data_to_load = "/Users/wheat/Pyber/data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df=pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
combined_df = pd.merge(ride_df,city_df,  on='city',how='left')

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#Using Loc method to get different type city list dataframe
urban_df = combined_df.loc[combined_df["type"] == "Urban", :]
suburban_df = combined_df.loc[combined_df["type"] == "Suburban", :]
rural_df = combined_df.loc[combined_df["type"] == "Rural", :]

#Groups our three new datasets by city
city_urban_df = urban_df.groupby(['city'])
city_suburban_df = suburban_df.groupby(['city'])
city_rural_df = rural_df.groupby(['city'])

#Get each city's ride count
urban_ride_count = city_urban_df['city'].value_counts()
suburban_ride_count = city_suburban_df['city'].value_counts()
rural_ride_count = city_rural_df['city'].value_counts()

#Get each city's driver count
urban_city_driver_count = city_urban_df['driver_count'].mean()
suburban_city_driver_count = city_suburban_df['driver_count'].mean()
rural_city_driver_count = city_rural_df['driver_count'].mean()


#Get average fare
urban_avg_fare = city_urban_df['fare'].sum() / urban_ride_count
suburban_avg_fare = city_suburban_df['fare'].sum() / suburban_ride_count
rural_avg_fare = city_rural_df['fare'].sum() / rural_ride_count

# Build the scatter plots for each city types
legend_urban = plt.scatter(urban_ride_count, urban_avg_fare, marker="o", facecolors="lightcoral", edgecolors="black",
            s=urban_city_driver_count*5, alpha=.5)
legend_suburban = plt.scatter(suburban_ride_count, suburban_avg_fare, marker="o", facecolors="lightskyblue", edgecolors="black",
            s=suburban_city_driver_count*5, alpha=.5)
legend_rural = plt.scatter(rural_ride_count, rural_avg_fare, marker="o", facecolors="gold", edgecolors="black",
            s=rural_city_driver_count*5, alpha=.5)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data 2018")
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
plt.legend((legend_urban, legend_suburban, legend_rural),
           ('Urban', 'Suburban', 'Rural'),
           scatterpoints=1,
           loc='upper right',
           fontsize=8)

# Incorporate a text label regarding circle size
plt.text(.95, 0.5, "Note: \nCircle size correlates with driver count per city.", 
         fontsize=10, transform=plt.gcf().transFigure)

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/PyberGraph.png")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fare = urban_df['fare'].sum()
suburban_fare = suburban_df['fare'].sum()
rural_fare = rural_df['fare'].sum()

# Build Pie Chart
labels = ["Urban",  "Rural","Suburban"]
total_fare_size = [urban_fare,  rural_fare, suburban_fare]
colors = ["lightcoral",  "gold","lightskyblue"]

explode = (.1,0,0)

plt.pie(total_fare_size, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=285)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/PercentageOfTotalFaresByCityType.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_ride_count = urban_df['city'].count()
suburban_ride_count = suburban_df['city'].count()
pie_rides_rural_count = rural_df['city'].count()

# Build Pie Chart
total_ride_size = [urban_ride_count,  pie_rides_rural_count, suburban_ride_count]
labels = ["Urban", "Rural","Suburban"]
colors = ["lightcoral",  "gold","lightskyblue"]
explode = (.1,0,0)
plt.pie(total_ride_size, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=260)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/PctOfTotalRidesByCityType.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_type_group= city_df.groupby(['type'])
driver_count_citytype= city_type_group['driver_count'].sum()

# Build Pie Charts
labels = ["Rural","Suburban","Urban"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0,.1)
plt.pie(driver_count_citytype, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=160)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("/Users/wheat/Pyber/Images/TotalDriversbyCityType.png")

# Show Figure
plt.show()
```
